```python
import pandas as pd
import numpy as np
data = pd.read_excel('prices_multiasset_portfolio_1292016__Homework_Data_Set.xlsx')
print(data.columns)
```

```python
print(data)
```

```python
#Test 1: 计算histogram
#unnamed:1 = 'SPTR Index',依此类推

import matplotlib.pyplot as plt
import numpy

operate = np.array([1,4,7,10,13,16,19,22])
target = operate[:]+1
for i in operate:
    fig1 = plt.figure()
    targetname = 'Unnamed: '+str(i);
    d = data[targetname].values

    # An "interface" to matplotlib.axes.Axes.hist() method
    n, bins, patches = plt.hist(x=d, bins='auto', color='#0504aa',
                                alpha=0.7, rwidth=0.85)
    plt.grid(axis='y', alpha=0.75)
    plt.xlabel('Value')
    plt.ylabel('Price')
    plt.title(targetname)
    maxfreq = n.max()
    # Set a clean upper y-axis limit.
    plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```
